```python
# Drop unnecessary columns
df.drop(columns=['index', 'Patient Id'], inplace=True)
```

## Preprossecing

```python
X = df.drop('Level', axis=1)
y = df['Level']
```

####  Encoding Target


```python
le = LabelEncoder()
y = le.fit_transform(y)
```

#### Preprossecing Pipeline

```python
preprossecing_Steps = list()
preprossecing_Steps.append(("Scaler" , StandardScaler()))
Preprossecing_Pipeline = Pipeline(steps=preprossecing_Steps)
```

## Building Model

```python

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.15, stratify=y, random_state=88
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
models = list()
models.append(("LR" , LogisticRegression(max_iter=50)))
models.append(("SVM" , SVC(C=0.15,kernel='rbf', random_state=42)))
models.append(("RF" , RandomForestClassifier(n_estimators=82, max_depth=3,min_samples_split=2)))
```

```python
for model in models:
    steps = list()
    steps.append(("Preprocessing" , Preprossecing_Pipeline))
    steps.append(model)
    pipeline = Pipeline(steps = steps)
    pipeline.fit(X_train, y_train)
    acc = pipeline.predict(X_test)
    print(model[0])
    print("Train_accuracy" ,pipeline.score(X_train, y_train))
    print("-" * 10)
    print("Test_accuracy" , accuracy_score(acc, y_test))
    print("-" * 10)
    print("classification report" , classification_report(acc, y_test))
    print("-" * 20)
    print("\n")
```

## Final Model
